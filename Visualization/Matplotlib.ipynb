# Scatterplot

```python
x = np.array([1,2,3,4,5,6,7,8])
y = x

plt.figure(figsize = (2, 2))
plt.scatter(x, y, c ="pink",
            linewidths = 2,
            marker ="s",
            edgecolor ="green",
            s = 50)
# similar to plt.plot(x, y, '.'), but the underlying child objects in the axes are not Line2D
```

```python
plt.figure(figsize=(2,2))
plt.plot(x, y, '.')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend( ["title_1", "Title_2"],  ncol = 1 , loc = "upper left" ,bbox_to_anchor =(1, 1))
plt.title("Connected Scatterplot points with lines")
plt.plot(x, y)
```
